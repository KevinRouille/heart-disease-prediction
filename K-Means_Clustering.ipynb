# Importing libraries and datasets

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, classification_report
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score
from sklearn.cluster import KMeans
#import sklearn.metrics as sm
import pickle
```

```
with open('/content/drive/MyDrive/Heart Disease Prediction/Datasets/df_combined.pickle', 'rb') as f:
    df = pickle.load(f)
```

# Splitting the features from the class variable and normalizing the data

```
X, y = df.drop(columns=['num']).values, df['num'].values
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

# Performing K-Means clustering and comparing predictions

```
clustering  = KMeans(n_clusters = 2,random_state = 123)
clustering.fit(X)
```

```
color_theme = np.array(['green','red'])
color_theme2 = np.array(['red','green'])
red_patch = mpatches.Patch(color='red', label='num=1')
green_patch = mpatches.Patch(color='green', label='num=0')

plt.figure(figsize=(10,2))
plt.scatter(y=df.cp,x=df.chol,c=color_theme[y], alpha = 0.7, )
plt.title('Actual Classification')
plt.xlabel('Cholesterol level (mg/dl)')
plt.ylabel('Chest pain')
plt.legend(handles = [green_patch, red_patch], scatterpoints = 1)
plt.plot()

plt.figure(figsize=(10,2))
plt.scatter(y=df.cp,x=df.chol,c=color_theme2[clustering.labels_], alpha = 0.7)
plt.title('K-Means Clustering')
plt.xlabel('Cholesterol level (mg/dl)')
plt.ylabel('Chest pain')
plt.legend(handles = [green_patch, red_patch], scatterpoints = 1)
plt.plot()
```

```
plt.figure(figsize=(7,4))
plt.scatter(y=df.age,x=df.chol,c=color_theme[y], alpha = 0.7, )
plt.title('Actual Classification')
plt.xlabel('Cholesterol level (mg/dl)')
plt.ylabel('Age')
plt.legend(handles = [green_patch, red_patch], scatterpoints = 1)
plt.plot()

plt.figure(figsize=(7,4))
plt.scatter(y=df.age,x=df.chol,c=color_theme2[clustering.labels_], alpha = 0.7)
plt.title('K-Means Clustering')
plt.xlabel('Cholesterol level (mg/dl)')
plt.ylabel('Age')
plt.legend(handles = [green_patch, red_patch], scatterpoints = 1)
plt.plot()
```
